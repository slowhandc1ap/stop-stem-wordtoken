```python

reviews = []
for fileid in movie_reviews.fileids():
  tag, filename = fileid.split('/')
  reviews.append((tag,movie_reviews.raw(fileid)))

sample = pd.DataFrame(reviews , columns =['target','document'])

def tokenize(text):
  tokens = word_tokenize(text)
  filtered_tokens = [word.lower() for word in tokens if word.isalnum() and word.lower()]
  return filtered_tokens

sample['word_toke'] = sample['document'].apply(tokenize)
sample
# print(sample.head())
# df = pd.DataFrame(sample)
# df
```

##ข้อ 2
จงเขียนโปรแกรมเพื่อทำ stemming จากคำที่ได้จากคำที่ได้จาก column word_token โดยสร้างเป็น column ใหม่ชื่อ stemmed ด้วย PorterStemmer

```python
from nltk.stem import PorterStemmer ## ลดรูปคำ

sample = pd.DataFrame(reviews, columns=['target', 'document'])

def tokenize(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word.lower() for word in tokens if word.isalnum() and word.lower()]
    return filtered_tokens

sample['word_token'] = sample['document'].apply(tokenize)

def stem_tokens(tokens):
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens

sample['stemmed'] = sample['word_token'].apply(stem_tokens)
sample
```
